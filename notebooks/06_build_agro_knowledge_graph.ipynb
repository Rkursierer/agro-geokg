Integrate individual RDF datasets into one final knowledge graph

* load RDF files
* merge graphs
* check number of triples
* save final Turtle graph

```python
from rdflib import Graph, Namespace, Literal
from rdflib.namespace import RDF, RDFS, XSD
```

```python
# import data / file paths

BASE_DIR = Path.cwd().parent  

OUTPUT_DIR = BASE_DIR / "data" / "processed"

INPUT_FILE_RDF_TESTAREA = BASE_DIR / "data" / "processed" / "test_area_individuals_with_geom.ttl"

INPUT_FILE_RDF_NDVI_AGRAR = BASE_DIR / "data" / "processed" /  "agro_individuals_with_geom_and_ndvi.ttl"

INPUT_FILE_RDF_OSM = BASE_DIR / "data" / "processed" / "osm_individuals_with_geom.ttl"

OUTPUT_FILE = BASE_DIR / "data" / "processed" / "final_agro_graph.ttl"

```

```python
g = Graph()

g.parse(INPUT_FILE_RDF_OSM, format="turtle")
g.parse(INPUT_FILE_RDF_NDVI_AGRAR, format="turtle")
g.parse(INPUT_FILE_RDF_TESTAREA, format="turtle")
```

```python
# Print the triples of the integratet knowledge graph
print(len(g))
```

```python
# save the final knowledge graph

g.serialize(
    destination=OUTPUT_FILE,
    format="turtle"
)

print("Triples:", len(g))
print("Saved:", OUTPUT_FILE)
```
